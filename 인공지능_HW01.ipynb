```python
#191113_김혜근_인공지능HW01_1번
import numpy as np

data01 = np.random.randint(1, 50, size=(5, 5)) #1, 50 사이 정수 난수 생성
data01
```

```python
#각 행의 합 구하기
data01_row_sum = data01.sum(axis=1)
data01_row_sum
```

```python
#각 열의 평균 구하기
data01_col_mean = data01.mean(axis=0)
data01_col_mean
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

#각 행의 인덱스 = X축, 각 행의 합 = Y축
plt.plot(np.arange(len(data01_row_sum)), data01_row_sum, 'ro')
```

```python
#각 열의 인덱스 = X축, 각 열의 평균 = Y축
plt.plot(np.arange(len(data01_col_mean)), data01_col_mean, 'ro')
```

```python
#191113_김혜근_인공지능HW01_2번
import numpy as np
import pandas as pd

#covid.csv 읽어오기
data02 = pd.read_csv('covid.csv')
data02
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

#2020년, 2021년 월별 확진자 수 추출
data02_2020 = data02.iloc[0, 2:]
data02_2021 = data02.iloc[1, 2:]

plt.title('COVID-19')
plt.plot(data02_2020, label = "2020") #2020년 확진자 수 그래프
plt.plot(data02_2021, label = "2021") #2021년 확진자 수 그래프
plt.xlabel('Year')   #X축 라벨
plt.ylabel('Number') #Y축 라벨
plt.legend(loc='upper left') #범례 표시
plt.show() 
```

```python
#191113_김혜근_인공지능HW01_3번
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.1, 5.0, 0.1) #x범위 지정
y1 = np.log(x)  #함수 1
y2 = np.sqrt(x) #함수 2
y3 = x          #함수 3
y4 = x**2       #함수 4

#그래프 1
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.xticks(np.arange(6))

#그래프 2
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.xticks(np.arange(6))

#그래프 3
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.xticks(np.arange(6))

#그래프 4
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.xticks(np.arange(6))

plt.tight_layout() #그래프 간 간격 조절
plt.show()
```
